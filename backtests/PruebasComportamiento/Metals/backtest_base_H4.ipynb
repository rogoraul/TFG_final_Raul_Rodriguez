```python
from pathlib import Path
import sys

cwd = Path.cwd().resolve()
root = next((path for path in [cwd, *cwd.parents] if (path / 'backtests' / 'enbolsa' / 'backtest_pipeline.py').exists()), None)
if root is None:
    raise RuntimeError('No se encontro la raiz del proyecto con backtests/enbolsa/backtest_pipeline.py')
if str(root) not in sys.path:
    sys.path.insert(0, str(root))

from data.sql.sql_funcs import get_symbols_by_group_normalized
from backtests.enbolsa.backtest_loader import cargar_portfolio_multiactivo
from backtests.enbolsa.backtest_pipeline import ejecutar_comparativa, colorear_comparativa, generar_desgloses_resultado, extraer_trades_resultado, extraer_operaciones_resultado, generar_auditoria_riesgo, resumir_portfolio_cargado, extraer_ventana_trade
```

```python
from backtests.enbolsa.backtest_loader import clear_portfolio_cache
clear_portfolio_cache()
```

```python
symbols = get_symbols_by_group_normalized(['Metals'])['Metals']

portfolio = cargar_portfolio_multiactivo(
    symbols,
    timeframe_ltf='H4',
    timeframe_htf='D1',
    group_name='Metals',
    verbose=True,
)

resultado = ejecutar_comparativa(
    portfolio,
    timeframe_ltf='H4',
    timeframe_htf='D1',
    return_details=True,
)
display(colorear_comparativa(resultado['summary']))
display(colorear_comparativa(resultado['splits']))
```

```python
desgloses = generar_desgloses_resultado(resultado, top_n=8)

print('Top activos por estrategia')
display(colorear_comparativa(desgloses['top_activos']))

print('Peores activos por estrategia')
display(colorear_comparativa(desgloses['bottom_activos']))

print('Distribucion de salidas')
display(desgloses['salidas'])
```

```python
display(resumir_portfolio_cargado(portfolio).tail(15))
```
